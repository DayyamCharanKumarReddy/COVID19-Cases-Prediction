```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data (replace 'your_file.csv' with your actual file path)
file_path = r'C:\Users\chara\Desktop\Predictioneer\train_data.csv'
data = pd.read_csv(file_path)

# Check if the dataset loaded successfully
print(data.head())

# 1. **Bar Plot**: Top 10 Locations by Deaths
top_10_deaths = data.nlargest(10, 'deaths')
plt.figure(figsize=(10, 6))
plt.barh(top_10_deaths['latitude'].astype(str) + ", " + top_10_deaths['longitude'].astype(str),
         top_10_deaths['deaths'], color='skyblue')
plt.xlabel('Deaths')
plt.ylabel('Locations (Lat, Long)')
plt.title('Top 10 Locations by Deaths')
plt.savefig('top_10_locations_by_deaths.png')
plt.show()

# 2. **Histogram**: Distribution of Case Fatality Ratio
plt.figure(figsize=(10, 6))
plt.hist(data['case_fatality_ratio'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of Case Fatality Ratio')
plt.xlabel('Case Fatality Ratio (%)')
plt.ylabel('Frequency')
plt.savefig('case_fatality_ratio_distribution.png')
plt.show()

# 3. **Scatter Plot**: Deaths vs Case Fatality Ratio
plt.figure(figsize=(10, 6))
plt.scatter(data['deaths'], data['case_fatality_ratio'], c='green', alpha=0.7, edgecolor='k')
plt.title('Deaths vs Case Fatality Ratio')
plt.xlabel('Deaths')
plt.ylabel('Case Fatality Ratio (%)')
plt.grid(True)
plt.savefig('deaths_vs_case_fatality_ratio.png')
plt.show()

# 4. **Line Plot**: Cumulative Deaths by Latitude
data_sorted = data.sort_values(by='latitude')
data_sorted['cumulative_deaths'] = data_sorted['deaths'].cumsum()
plt.figure(figsize=(10, 6))
plt.plot(data_sorted['latitude'], data_sorted['cumulative_deaths'], marker='o', color='red')
plt.title('Cumulative Deaths by Latitude')
plt.xlabel('Latitude')
plt.ylabel('Cumulative Deaths')
plt.grid(True)
plt.savefig('cumulative_deaths_by_latitude.png')
plt.show()
```
